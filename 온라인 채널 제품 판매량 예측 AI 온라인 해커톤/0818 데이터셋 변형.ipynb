#train 데이터셋 변형

```python
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LGaimers/data/train.csv', encoding='cp949')
train
```

```python
train.info()
```

```python
train.isna().sum()
```

```python
df1 = train.drop(columns = ['제품', '대분류', '중분류', '소분류', '브랜드']); df1
```

```python
df1 = df1.melt('ID'); df1
```

```python
df1.sort_values(['ID', 'variable'], ascending=True, inplace=True); df1
```

```python
df1.set_index('variable', inplace=True); df1
```

```python
df1 = df1.rename(columns = {'value': 'count'}); df1
```

```python
df1.to_csv('/content/drive/MyDrive/Colab Notebooks/LGaimers/8.18_onlytrain.csv')
```

#sales 데이터셋 변형

```python
sales = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LGaimers/data/sales.csv', encoding='cp949'); sales
```

```python
df2 = sales.drop(columns = ['제품', '대분류', '중분류', '소분류', '브랜드'])
df2 = df2.melt('ID')
df2.sort_values(['ID', 'variable'], ascending=True, inplace=True)
df2.set_index('variable', inplace=True); df2
```

```python
df2 = df2.rename(columns = {'value': 'sales'}); df2
```

```python
df2.to_csv('/content/drive/MyDrive/Colab Notebooks/LGaimers/8.18_onlysales.csv')
```
